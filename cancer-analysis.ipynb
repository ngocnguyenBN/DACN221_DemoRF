# Random forest Classifier

```python
df.diagnosis = [1 if i == "M" else 0 for i in df.diagnosis]
```

```python
x = df.drop(["diagnosis"], axis = 1)
y = df.diagnosis.values
```

```python
x = (x - np.min(x)) / (np.max(x) - np.min(x))
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 42)
```

```python
print('X train shape: ', x_train.shape)
print('Y train shape: ', y_train.shape)
print('X test shape: ', x_test.shape)
print('Y test shape: ', y_test.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

rf = RandomForestClassifier(n_estimators = 100 , random_state = 1)
rf_train = rf.fit(x_train, y_train)
y_pred_rf = rf_train.predict(x_test)

print("Random forest algorithm accuracy result:" , rf.score(x_test, y_test))
print('Accuracy Score RF:  {:.4f}'.format(accuracy_score(y_test, y_pred_rf)))

```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred_rf))
```

```python
rfc = RandomForestClassifier(random_state=42)

n_estimators = [int(x) for x in np.linspace(start = 200, stop = 2000, num = 10)]
max_features = ['auto', 'sqrt', 'log2']
max_depth = [int(x) for x in np.linspace(10, 110, num = 11)]
max_depth.append(None)
min_samples_split = [2, 5, 10]
min_samples_leaf = [1, 2, 4]
bootstrap = [True, False]

params_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf,
               'bootstrap': bootstrap}
```

```python
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
rfc_cv = RandomizedSearchCV(estimator = rfc,  param_distributions = params_grid, verbose = 0, cv = 5, n_iter = 100)

rfc_cv.fit(x_train, y_train)
```

```python
print(f'best: {rfc_cv.best_params_}')
```

```python
rf_cv = RandomForestClassifier(n_estimators = 1600, min_samples_split = 2, min_samples_leaf = 2, max_features = 'log2', max_depth = 20, bootstrap = False, random_state = 42)

modelrf_cv = rf_cv.fit(x_train, y_train)

y_pred_rf_cv = modelrf_cv.predict(x_test)
```

```python
print('Accuracy Score RF + RandomSearchCV: {:.4f}'.format(accuracy_score(y_test, y_pred_rf_cv)))
```

```python
print(classification_report(y_test, y_pred_rf_cv))
```
